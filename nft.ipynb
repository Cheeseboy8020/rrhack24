```python
from solcx import install_solc, compile_files

install_solc("0.6.2")
```

```python
compiled_sol = compile_files(
    ["./MyNFT.sol"],
    output_values=['abi', 'bin'],
    solc_version="0.6.2"
)

compiled_sol
```

```python
# get bytecode / bin
bytecode = compiled_sol["./MyNFT.sol:MyNFT"]['bin']

# get abi
abi = compiled_sol["./MyNFT.sol:MyNFT"]['abi']
```

```python
import os
from web3 import Web3

w3 = Web3(Web3.HTTPProvider("http://127.0.0.1:7545"))


private_key = os.environ["pk"]
account = w3.eth.account.from_key(private_key).address
account
```

```python
contract = w3.eth.contract(abi=abi, bytecode=bytecode)
tx_hash = contract.constructor().transact({"from": account})
tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)
contract_address = tx_receipt.contractAddress
```

```python
nft = w3.eth.contract(address=contract_address, abi=abi)

nft.functions.balanceOf(account).call()
```

```python
nft.functions.mintNFT(account, "Granny Smith").transact({"from": account})
```

```python
nft.functions.tokenURI(3).call()
```

```python
nft.functions.transferFrom(account, "0x8d883Bc4D58FD92920f603D727A1d371B6701112",3).transact({"from": account})
```
